1. pass additionalparameter as explode and define its size

#### Colors and Legends for a pie chart

```python
food = np.array([10, 25, 20, 15, 22])
put_labels = ["MILK", "BUTTER", "BREAD", "TEA", 'COFFEE']
put_explode = [0.2,0,0,0.1,0]
put_colors = ["g", "k", "c", "r","y"] ## k is for black
plt.pie(food, labels = put_labels, 
              startangle = 120, 
              explode = put_explode, 
              shadow = True, 
              colors = put_colors, 
              radius = 1)
plt.legend(title = "FOOD"); ### for changing the position of the label
```

Parameters of PIE PLOT
    
    startangle = start angle of the pie 
    explode = to explode a particular portion of the pie 
    shadow = to put the shadow 
    colors = for the color of the pie 
    autopct= to display the percent value using Python string formatting ### autopercent
    pctdistance = to set the distance of autopct from center 
    labeldistance = to set the label distance 
    counterclock = direction of the plot
    radius = size of the plot, 
    rotatelabels = to rotate the label according to it's correspoding pie
        wedgeprops = Dict of arguments passed to the wedge objects making the pie. 
    For example, you can pass in wedgeprops = {'linewidth': 3} to set the width of the wedge border lines equal to 3

    LEGEND
    prop = {"size":10} - Legend Font Size
    title = "FOOD" - title of the legend
    loc= "upper right" - Allignment of the legend 
    bbox_to_anchor=(2.5, -0.15) - coorinates of the legend -(minus) specifies that legend wll go to bottom.
    ncol=3 - number of columns in the legend

```python
food = np.array([10, 25, 20, 15, 22])
put_labels = ["MILK", "BUTTER", "BREAD", "TEA", 'COFFEE']
put_explode = [0.2,0,0,0,0]

put_colors = ["g", "b", "c", "r","y"]

plt.pie(food, 
        labels = put_labels, 
        startangle = 90, 
        explode = put_explode, 
        shadow = True, 
        colors = put_colors, 
        autopct='%1.2f%%', 
        pctdistance = 0.8, 
        labeldistance = 1.2, 
        counterclock = False,
        radius = 2, 
        rotatelabels = True,
        wedgeprops=dict(width=1));

plt.legend(prop = {"size":10}, title = "FOOD", loc="upper left", bbox_to_anchor=(2.5, -0.15));
```

### SUBPLOTS
1. used when there are more than one graphs
2. plt.subplot()

### Plottin two graphs side by side using subplot(x,y,z)
    x - number of rows
    y - numbe of columns
    z - postion of the plot
    For Example: 
    plt.subplot(1,2,1) specifies that the plot has 1 row, 2 column, and graph will be placed at the first position
    plt.subplot(1,2,2) specifies that the plot has 1 row, 2 column, and 2nd plot

```python
## plot 1
x1 = np.array([0, 1, 2, 3])
y1 = np.array([2, 4, 7, 9])

plt.subplot(1, 2, 2)### Can also be written as plt.subplot(121)
plt.plot(x1,y1)

#plot 2:
x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 30, 19, 20])

plt.subplot(1, 2, 1)### Can also be written as plt.subplot(122)
plt.plot(x2,y2)
```

### Example 2

```python
'''plotting graph one over the other:
here, plt.subplot(2,1,1) specifies that the plot has 2 row, 1 column, and 1st plot and 
plt.subplot(2,1,2) specifies that the plot has 2 row, 1 column, and 2nd plot.'''

x1 = np.array([0, 1, 2, 3])
y1 = np.array([2, 4, 7, 9])


plt.subplot(2, 1, 1)
plt.plot(x1,y1)

#plot 2:
x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 30, 19, 20])

plt.subplot(2, 1, 2)
plt.plot(x2,y2)
```

### Creating multiplots in the same graph
1. Make use of subplot function
2. specify the number of rows and number of columns and also thesize of the figures
3. On the LHS specify fig (for figure) and ax (for the number and types of graphs to be plotted) 

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10, 10))

ax1.plot(x1,y1)
ax1.set(title="Line Chart", xlabel="x-axis", ylabel="y-axis")

ax2.bar(items.keys(),items.values())
ax2.set(title="Bar Chart", xlabel="x-axis", ylabel="y-axis")

ax3.pie(food, labels = put_labels)
ax3.set(title="Pie Chart", xlabel="x-axis", ylabel="y-axis")

ax4.barh(list(items.keys()),list(items.values()),color = 'g')
ax4.set(title="Horizontal Bar Chart", xlabel="x-axis", ylabel="y-axis")
```

### Overlapping

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

ax[0,0].plot(x1,y1,c='r')
ax[0,0].set(title="Line Chart", xlabel="x-axis", ylabel="y-axis")

ax[0,0].bar(items.keys(),items.values())
ax[0,0].set(title="Bar Chart", xlabel="x-axis", ylabel="y-axis")

ax[1,0].pie(food, labels = put_labels)
ax[1,0].set(title="Pie Chart", xlabel="x-axis", ylabel="y-axis")

ax[1,1].barh(list(items.keys()),list(items.values()),color = 'g')
ax[1,1].set(title="Horizontal Bar Chart", xlabel="x-axis", ylabel="y-axis")

ax[0,1].bar(items.keys(),items.values())
ax[0,1].set(title="Bar Chart", xlabel="x-axis", ylabel="y-axis")
```

#### Annotations over Bar Plot

```python
'''x-coordinate: bar.get_x() + bar.get_width() / 2 
y-coordinate: bar.get_height() 
free space to be left to make graph pleasing: (0, 8) 
ha and va stand for the horizontal and vertical alignment'''

food = np.array([10, 25, 20, 15, 22])
put_labels = ["MILK", "BUTTER", "BREAD", "TEA", 'COFFEE']

x = np.arange(len(put_labels))  # the label locations
width = 0.5  # the width of the bars 

plots = plt.bar(x,food)
plt.xticks(x, labels = put_labels)

#iterate over every bar
for bar in plots:
    plt.annotate(format(bar.get_height(), '.2f'),  
                   (bar.get_x() + bar.get_width() / 2,  
                    bar.get_height()), ha='center', va='center', 
                   size=10, xytext=(0,5), 
                   textcoords='offset points')
```

#### Marking the points

```python
'''Scatter Plot'''
#hatch = to set the pattern over the bar

sample.plot(x ='Age', y = 'Salary', kind = "scatter", hatch = '//', s= 100 )

for i, txt in enumerate(sample['Salary']):
    plt.annotate(txt, (sample['Age'][i], sample['Salary'][i]))
```

```python
import math
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a


x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

# Move left y-axis and bottim x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x,sig)
plt.show()
```
