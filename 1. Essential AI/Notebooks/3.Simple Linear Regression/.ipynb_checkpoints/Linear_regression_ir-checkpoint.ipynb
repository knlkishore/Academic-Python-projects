```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ir = pd.read_csv("lr_diabetes.csv")
ir.shape
```

```python
ir.head()
```

```python
ir.describe()
```

```python
#Plotting the data
ir.plot(x='bmi', y='Progression', style='o')
plt.title('bmi vs Progression')
plt.xlabel('bmi')
plt.ylabel('Progression')
plt.show()
```

```python
#preparing the data
X = ir.iloc[:, :-1].values
y = ir.iloc[:, 1].values


```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

reg.fit(X, y)
```

```python
#printing parameters:
print(reg.intercept_)
print(reg.coef_)
```

```python
#making predictions:
y_pred = reg.predict(X)
```

```python
df = pd.DataFrame({'Actual': y, 'Predicted': y_pred})
df
```

```python
# The coefficients
from sklearn.metrics import mean_squared_error, r2_score
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('R^2 Score: %.2f'
      % r2_score(y, y_pred))
```

```python
# Plot outputs
plt.scatter(X, y,  color='black')
plt.plot(X, y_pred, color='blue', linewidth=2)


plt.xticks(())
plt.yticks(())

plt.show()
```

Lasso Regression


```python
from sklearn.linear_model import Lasso
reg1 = Lasso(alpha=1.0)

reg1.fit(X, y)
```

```python
#printing parameters:
print(reg1.intercept_)
print(reg1.coef_)
```

```python
#making predictions:
y_predsvm = reg1.predict(X)
```

```python
df1 = pd.DataFrame({'Actual': y, 'Predicted': y_predsvm})
df1
```

```python
# The coefficients
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y, y_predsvm))

# The coefficient of determination: 1 is perfect prediction
print('R^2 Score: %.2f'
      % r2_score(y, y_predsvm))
```
